# Fitting lines

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot")
plt.rcParams["figure.figsize"]=(18,12)
```

```python
x = np.linspace(0.0,10.0,100)
```

```python
y = 3.0*x+2.0
```

```python
plt.plot(x,y,'o')
```

```python
np.polyfit(x,y,1)
```

```python
y= 3.0*x+2.0 + np.random.normal(0.0,2,len(x))
```

```python
plt.plot(x,y,'o')
```

```python
coeffs = np.polyfit(x,y,1)
coeffs
```

```python
plt.plot(x,y,'.', label = "data")
plt.plot(x, coeffs[0] * x + coeffs[1],'-', label="Best fit")
plt.legend()
```

```python
x1 =np.linspace(0.0,0.5,20)
```

```python
y1 =3.0*x1+2.0+np.random.normal(0.0,1.0,len(x1))
```

```python
x2 =np.linspace(0.0,0.5,30)
```

```python
y2 =6.0* x2 +5.0 + np.random.normal(0.0,1.0,len(x2))
```

```python
coeffs1 = np.polyfit(x1,y1,1)
coeffs1
```

```python
coeffs2 = np.polyfit(x2,y2,1)
coeffs2
```

```python
plt.plot(x1,y1,'.', label = "data set 1")
plt.plot(x1, coeffs1[0] * x1 + coeffs1[1],'-', label="Best fitline 1")
plt.legend()
plt.plot(x2,y2,'.', label = "data set 2")
plt.plot(x2, coeffs2[0] * x2 + coeffs2[1],'-', label="Best fitline 2")
plt.legend()
```

```python
x = np.concatenate([x1,x2])
y = np.concatenate([y1,y2])
```

```python
coeffs =np.polyfit(x,y,1)
coeffs
```

```python
plt.plot(x1,y1,'r.', label = "data set 1")
plt.plot(x1, coeffs1[0] * x1 + coeffs1[1],'o', label="Best fitline 1")
plt.legend()
plt.plot(x2,y2,'b.', label = "data set 2")
plt.plot(x2, coeffs2[0] * x2 + coeffs2[1],'x', label="Best fitline 2")
plt.legend()
plt.plot(x,y,'.', label = "combined data set")
plt.plot(x, coeffs[0] * x + coeffs[1],'-', label="Best combined")
plt.legend()
```

```python
flipper_length_mm = ['flipper_length_mm']
body_mass_g = ['body_mass_g']

flipper_train = flipper_length_mm[:-20]
flipper_test = flipper_length_mm[-20:]

mass_train = body_mass_g[:-20]
mass_test = body_mass_g[-20:]
```

```python
# Create linear regression object
regr = linear_model.LinearRegression()

flipper_length_mm = flipper_length_mm[:, np.newaxis, 1]

# Train the model using the training sets
regr.fit(flipper_train, mass_train)

# Make predictions using the testing set
mass_pred = regr.predict(mass_test)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(mass_test, mass_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(mass_test, mass_pred))

# Plot outputs
plt.scatter(flipper_test, mass_test,  color='black')
plt.plot(flipper_test, mass_pred, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```

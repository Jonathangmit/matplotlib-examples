##### Normal

https://numpy.org/doc/stable/reference/random/generated/numpy.random.Generator.normal.html

```python
rng.normal()
```

```python
nos2 = rng.normal(20,3,size=100)
nos2
```

```python
plt.hist(nos2,bins=20)
```

```python
nos_A = rng.normal(1.0,size=1000)
nos_B = rng.normal(3.0,size=1000)
```

```python

fig, ax = plt.subplots()
ax.hist(nos_A, bins=100,alpha=0.5)
ax.hist(nos_B, bins=100,alpha=0.5);
```

```python
nos_A = rng.normal(scale=0.5,size=1000)
nos_B = rng.normal(scale=0.1,size=1000)
```

```python
fig, ax  = plt.subplots()
ax.hist(nos_A, bins=100,alpha=0.5,density = True)
ax.hist(nos_B, bins=100,alpha=0.5,density = True);
```

###### Normal samples

```python
# Generate 10,000 samples of 100 normally generated numbers
samples = rng.normal(loc=12.34,size=(10000,100))
```

```python
samples
```

```python
samples.shape
```

```python
plt.hist(samples.flatten(), bins=20);
```

```python
# calculate the mean of each sample

samples.mean()
```

```python
mean_samples = samples.mean(axis=1)
```

```python
mean_samples
```

```python
mean_samples.shape
```

```python
plt.hist(mean_samples, bins=20);
```

```python
samplesA = rng.normal(loc=1.0,scale=2.0,size=(10000,100))
samplesB = rng.normal(loc=1.1,scale=2.0,size=(10000,100))
```

```python
#Calulate the means of th samples in meach set
```

```python
mean_samplesA = samplesA.mean(axis=1)
mean_samplesB = samplesB.mean(axis=1)
```

```python
fig, ax =plt.subplots()
ax.hist(mean_samplesA, alpha =0.5, bins=40, density=True)
ax.hist(mean_samplesB, alpha =0.5, bins=40, density=True);
```

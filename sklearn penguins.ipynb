```python
# Flipper length v Body mass
data_columns =(penguins7['flipper_length_mm'])
target_column=(penguins7['body_mass_g'])
sns.scatterplot(data=penguins, x=data_columns, y=target_column)
```

```python
data_columns_train,data_columns_test,target_column_train,target_column_test=train_test_split(data_columns,target_column,train_size =50,test_size=283)
```

```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(data_columns_train,target_column_train )
```

```python
target_column_prediction=regr.predict(data_columns_test)
```

```python
print('Coefficients: \n', regr.coef_)
```

```python
print('Mean squared error: %.2f'
% mean_squared_error(target_column_test,target_column_prediction ))
```

```python
# Plot outputs
plt.style.use('ggplot')
plt.rcParams['figure.figsize']=14,10
plt.scatter(data_columns_test,target_column_test ,  color='black')
plt.plot(data_columns_test,target_column_prediction , color='blue', linewidth=3)
plt.xlabel('flipper length in mm')
plt.ylabel('body mass in grams')
plt.title('linear regression using ')

plt.xticks()
plt.yticks()

plt.show()
```
